# Plotting

```python
caseinfo_batch0_count = caseinfo_batch0[caseinfo_batch0['ProvRes']== 'NCR'].groupby('DateRepConf').count().CaseCode
caseinfo_batch1_count = caseinfo_batch1[caseinfo_batch0['ProvRes']== 'NCR'].groupby('DateRepConf').count().CaseCode
caseinfo_batch2_count = caseinfo_batch2[caseinfo_batch0['ProvRes']== 'NCR'].groupby('DateRepConf').count().CaseCode.drop('2021-09-01')

df = pd.concat([caseinfo_batch0_count, caseinfo_batch1_count, caseinfo_batch2_count])
```

```python
fig, ax = plt.subplots(figsize=(10, 10))

df.plot()
```

```python
dates = df.index.tolist()
days_since = [i for i in range(1, 601)]
cases = df.tolist()
```

```python
len(dates)
```

```python
from scipy.optimize import curve_fit

fig, ax = plt.subplots(figsize=(10,10))

def objective_poly3(x, a, b, c, d):
    return a*x + b*x**2 + c*x**3 + d
def objective_poly2(x, a, b, c):
    return a*x + b*x**2 + c
def objective_linear(x, m, b):
    return m*x + b
def objective_exp(x, a, b):
    return a*(b**x)
    
x = np.linspace(1, 605)

popt, _ = curve_fit(objective_poly2, days_since, cases)
a, b, c = popt 
print('y = %.5f * x + %.5f * x**2 + %.5f' % (a, b, c))
y = objective_poly2(x, a, b, c)
plt.plot(x, y, label='2nd Order Polynomial Fit', color='blue')

popt, _ = curve_fit(objective_poly3, days_since, cases)
a, b, c, d = popt
y = objective_poly3(x, a, b, c, d)
plt.plot(x, y, label='3rd Order Polynomial Fit', color='green')

popt, _ = curve_fit(objective_linear, days_since, cases)
m, b = popt
y = objective_linear(x, m, b)
plt.plot(x, y, label='Linear Fit', color='purple')

popt, _ = curve_fit(objective_exp, days_since, cases)
a, b = popt
y = objective_exp(x, a, b)
plt.plot(x, y, label='Exponential Fit', color='red')

plt.plot(days_since, cases, color='k', alpha=0.5)

# ECQ is Red
# MECQ is Pink
# MGCQ is Green
# GCQ is Yellow
plt.axvspan(7, 68, facecolor='red', alpha=0.2)
plt.axvspan(68, 84, facecolor='pink', alpha=0.2)
plt.axvspan(84, 129, facecolor='yellow', alpha=0.2)
plt.axvspan(129, 163, facecolor='pink', alpha=0.2)
plt.axvspan(163, 386, facecolor='yellow', alpha=0.2)
plt.axvspan(386, 399, facecolor='red', alpha=0.2)
plt.axvspan(399, 432, facecolor='pink', alpha=0.2)
plt.axvspan(432, 515, facecolor='yellow', alpha=0.2)
plt.axvspan(515, 530, facecolor='red', alpha=0.2)
plt.axvspan(530, 556, facecolor='pink', alpha=0.2)
plt.axvspan(556, 620, facecolor='purple', alpha=0.2)


plt.legend()
plt.xlabel('Days since the First COVID Case')
plt.ylabel('Number of Daily COVID Cases')
```
